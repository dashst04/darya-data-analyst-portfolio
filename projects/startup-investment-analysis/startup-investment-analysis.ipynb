Перед тем как определить, что мы можем сделать с пропусками в этих столбцах, ознакомимся с количеством пропуском в абсолютном и относительном значениях.

```python
# Для вывода пропущенных значений создадим функцию show_missing_stats, которая отобразит количество пропусков в абсолютном и относительном значениях
def show_missing_stats(tmp0):
    missing_stats = pd.DataFrame({
        'Кол-во пропусков': tmp0.isnull().sum(),
        'Доля пропусков': tmp0.isnull().mean()
    })
    missing_stats = missing_stats[missing_stats['Кол-во пропусков'] > 0]
    
    if missing_stats.empty:
        return "Пропусков в данных нет"
    
# Форматируем при выводе через Styler
    return (missing_stats.style.format({'Доля пропусков': '{:.6f}'}))
show_missing_stats(df)
```

Во всех столбцах, кроме `first_funding_at`, количество пропущенных значений составляет более 20%, поэтому:
* строки с пропусками в `first_funding_at` удалим, так их всего 2 и это не сильно исказит анализ.

### 2.1. Группы по срокам финансирования

На данном этапе, разделим все компании на 3 группы: 
    
* Единичное финансирование — был всего один раунд финансирования.

* Срок финансирования до года — между первым и последним раундом финансирования прошло не более года.

* Срок финансирования более года.

Для начала, сделаем так, чтобы для всех визуализаций была использована единая цветовая схема:
* единичное финансирование — розовый цвет,
* срок финансирования до года — оранжевый,
* срок финансирования более года — жёлтый.

```python
# Создадим словарь цветов:
colors = {
    'Единичное финансирование': '#fc0fc0',
    'Срок финансирования до года': '#ff7f0e',
    'Срок финансирования более года': '#ffba00'
}
```

```python
# Создадим функцию, чтобы разделить компании на группы по срокам финансирования
def funding_group(row):
    if row['funding_rounds'] == 1:
        return 'Единичное финансирование'
    
    duration = (row['last_funding_at'] - row['first_funding_at']).days
    
    if duration <= 365:
        return 'Срок финансирования до года'
    else:
        return 'Срок финансирования более года'

# Создадим столбец 'funding_group' и применим функцию к датафрейму:
df['funding_group'] = df.apply(funding_group,axis=1)

# Проверим результат
df.head(5)
```

В рамках инжиниринга признаков для каждой компании был создан новый столбец `funding_group`, отражающий срок финансирования. Категория определялась на основе количества раундов и длительности периода между первым и последним финансированием.

* По объёму инвестиций: Отобразите, какую долю от общего объёма привлечённых средств получила каждая группа.

Сначала, создадим график соотношения групп финансирования по количеству компаний.

```python
# Посчитаем количество компаний для каждой категории в относительном значении:
group_counts = df['funding_group'].value_counts(normalize=True)

# Выведем результат
group_counts
```

Построим график соотношения групп финансирования по объему инвестиций.

* сегменты, в которые входит от 35 до 120 включительно, отнесем к средним; 
* сегменты до 35 компаний отнесем к нишевым. 

Для каждой из категорий рассчитаем, сколько сегментов попадает в каждую из них.

Для дальнейшего анализа оставим в столбце `market` только массовые сегменты. Для остальных сегментов заменим значения на заглушки `niche` для нишевых и `mid` для средних.

## Шаг 3. Работа с выбросами и анализ

### 3.1. Анализируем и помечаем выбросы в каждом из сегментов

Нас интересует обычный для рассматриваемого периода размер средств, который предоставлялся компаниям.

Построим гистограмму и диаграмму размаха для столбца `funding_total_usd` чтобы оценить, какой размер общего финансирования для одной компании будет типичным, а какой — выбивающимся.

С помощью IQR найдем интервал, в котором лежат типичные значения.

```python
# Считаем Q1 и Q3
Q1 = df['funding_total_usd'].quantile(0.25)
Q3 = df['funding_total_usd'].quantile(0.75)

# Вычисляем IQR
IQR = Q3 - Q1

# Определяем верхнюю и нижнюю границы нормальных значений
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Выводим результат
print(f'Типичный размер финансирования находится в интервале {Q1} - {Q3}')
print(f'Аномальные значения: ниже {lower_bound} или выше {upper_bound}')
```

По boxplot-графику видно, что распределение общего объёма финансирования сильно скошено, и сделать какие либо выводы по графику будет трудно. Поэтому, мы посчитали межквартильный интервал (IQR). 
Согласно IQR:

Типичные значения общего финансирования одной компании лежат в интервале между первым и третьим квартилями — от 350000.0 до 10000000.0 долларов.

Значения, превышающие верхнюю границу межквартильного размаха (24475000.0 долларов), можно считать аномально высокими и относящимися к выбросам.

Определим компании с аномальным объёмом общего финансирования - применим метод IQR отдельно по каждому сегменту.

```python
# Для каждого сегмента считаем Q1 и Q3
def mark_outliers(group):
    Q1 = group['funding_total_usd'].quantile(0.25)
    Q3 = group['funding_total_usd'].quantile(0.75)
    # Считаем IQR
    IQR = Q3 - Q1

    # Определяем нижнию и верхнию границы  
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Помечаем выбросы
    group['is_outlier'] = ((group['funding_total_usd'] < lower_bound) | (group['funding_total_usd'] > upper_bound))
    
    return group

# Применяем по сегментам:
df_outliers = df.groupby('market', group_keys=False).apply(mark_outliers)
```

Определим сегменты рынка с наибольшей долей компаний, получивших аномальное финансирование, и выведем топ таких сегментов.

```python
# Рассчитываем для каждого сегмента долю компаний, получивших аномальное финансирование и сортируем в порядке убывания
outlier_stats = (df_outliers.groupby('market')['is_outlier'].mean().sort_values(ascending=False))

# Выводим топ-10 сегментов рынка
outlier_stats.head(10)
```

Наибольшая доля компаний с аномальным объёмом общего финансирования наблюдается в сегменте Real Estate - около 18%. В отсальных сегментах доля выбросов находится в диапазоне 15–16%. Это говорит о том, что в данных сегментах чаще встречаются компании, привлёкшие существенно больший объём инвестиций по сравнению с типичными значениями внутри своего рынка.

### 4.3 Годовая динамика доли возвращённых средств по типам финансирования

Нам необходимо для каждого года и каждого вида финансирования рассчитать нормированные значения возврата средств: то есть какую долю возвращённые средства составляют от предоставленных. При этом слишком большие аномальные значения, то есть неадекватные выбросы, нужно заменить на пропуски.

```python
# Считаем суммарный объём привлечённого финансирования по годам и типам
# Список типов финансирования
funding_types_2 = ['venture','debt_financing','private_equity','seed','angel']

# Считаем суммарный объем привлеченных средств по годам
funding_by_year = (df_final.groupby('funding_year')[funding_types_2].sum().reset_index().rename(columns={'funding_year': 'year'}))
```

```python
# Приводим возвраты к тем же единицам измерения (USD):
# Копируем таблицу возвратов
returns = cb_returns[funding_types].copy()

# Переводим миллионы долларов в доллары
returns[funding_types] = returns[funding_types] * 1_000_000

# Переносим индекс в столбец для объединения
returns = returns.reset_index().rename(columns={'index': 'year'})
```

```python
# Объединяем инвестиции и возвраты по годам
merged_df = funding_by_year.merge(returns,on='year',suffixes=('_funding', '_returns'))
```

```python
# Считаем нормированную долю возврата средств (с защитой от деления на ноль)
# Малое число для защиты от деления на ноль
for ftype in funding_types_2:
    merged_df[f'{ftype}_ratio'] = (merged_df[f'{ftype}_returns'] / (merged_df[f'{ftype}_funding'] + 1e-60))
```

```python
# Убираем аномальные выбросы (слишком большие значения)
# Используем IQR-метод отдельно для каждого типа финансирования:
for ftype in funding_types_2:
    ratio_col = f'{ftype}_ratio'
    
    Q1 = merged_df[ratio_col].quantile(0.25)
    Q3 = merged_df[ratio_col].quantile(0.75)
    IQR = Q3 - Q1
    
    upper_bound = Q3 + 1.5 * IQR
    
    # Заменяем аномалии на NaN
    merged_df.loc[merged_df[ratio_col] > upper_bound, ratio_col] = None
```

Построим график, на котором отобразим нормированные значения возврата средств для типов финансирования `venture`, `debt_financing`, `private_equity`, `seed` и `angel`.

```python
# Строим график нормированных значений возврата средств
plt.figure(figsize=(12, 8))

for ftype in funding_types_2:
    plt.plot(
        merged_df['year'],
        merged_df[f'{ftype}_ratio'],
        marker='o',
        label=ftype)

# Добавляем заголовок графика и подписи для осей
plt.title('Годовая динамика доли возвращённых средств по типам финансирования')
plt.xlabel('Год')
plt.ylabel('Доля возвращённых средств')

# Добавляем легенду и сетку
plt.legend(title='Тип финансирования')
plt.grid(True)

plt.show()
```

Анализ годовой динамики доли возвращённых средств показывает, что наиболее устойчивый рост показателя наблюдается в сегменте private equity: начиная с середины рассматриваемого периода, доля возвратов постепенно увеличивается и достигает максимальных значений к концу периода. 

Angel-финансирование демонстрирует умеренный рост в отдельные годы, но без устойчивой положительной динамики на всём временном интервале.

2. Проанализирована динамика типичного размера раунда и общего количества раундов по годам;

3. Установлено, что максимальный типичный размер одного раунда наблюдался в 2013 году;

4. В 2014 году наблюдался рост среднего размера финансирования и снижение количества раундов.

5. Проведён анализ динамики суммарного объёма финансирования по массовым сегментам рынка.

6. Для ключевых типов финансирования (venture, debt_financing, private_equity, seed, angel) была рассчитана годовая доля возвращённых средств.

#### Выводы и рекомендации

На основе проведённого анализа можно рекомендовать в 2015 году инвестировать в отрасль Technology, так как данный сегмент демонстрирует устойчивый рост объёмов финансирования и высокий интерес со стороны инвесторов.

Наиболее уместным типом финансирования для вложений являются инвестиции в виде прямых вложений, поскольку он характеризуется сравнительно стабильной и растущей долей возвращённых средств, что снижает инвестиционные риски.

В целом результаты анализа согласуются между собой: сегменты с ростом инвестиционной активности, как правило, сопровождаются более устойчивыми показателями возвратности.
